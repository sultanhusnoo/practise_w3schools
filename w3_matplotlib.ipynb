```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,360)
y1 = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))
y3 = np.tan(np.radians(x))
```

```python
len(y1)
```

```python
plt.plot(x,y1,label="sin(x)")
plt.xlabel("Degrees")
plt.ylabel("Sine Function")
plt.legend()
```

```python
plt.plot(x,y2)
```

```python
plt.plot(x,y3)
plt.ylim(-10,10)
```

```python
plt.plot(x,y1,color='red',label="sin(x)")
plt.plot(x,y2,color='blue',label="cos(x)")
plt.plot(x,y3,color='black',label="tan(x)")
plt.ylim(-2.5,2.5)
plt.grid()
plt.xlabel("x / degrees")
plt.ylabel("Trig Fn")
plt.legend()
```

```python
x1 = np.arange(0,360)
y_1 = np.sin(np.radians(x))
y_2 = np.sin(2* np.radians(x))
y_3 = np.sin(3* np.radians(x))
```

```python
plt.plot(x1,y_1,color='red',label="sin(x)")
plt.plot(x1,y_2,color='blue',label="sin(2x)")
plt.plot(x1,y_3,color='black',label="sin(3x)")
plt.grid()
plt.xlabel('Degrees')
plt.ylabel('Sine Functions')
plt.legend()
```

```python
x3 = np.arange(0,360,10)
y_scatter = np.sin(np.radians(x3))
plt.scatter(x3,y_scatter,marker='.')
```

```python
x4 = np.arange(0,360,15)
y_a=np.sin(np.radians(x4))
plt.vlines(x4,0,y_a,color='black')
plt.scatter(x4,y_a,marker='*')
plt.grid()
```
